## 1. Numbers & Math

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `abs(x)` | `x`: a number (int, float, or complex). Returns its magnitude. |
| `divmod(a, b)` | `a`, `b`: numbers. Returns the tuple `(a // b, a % b)`. |
| `pow(base, exp, mod=None)` | `base`, `exp`: numbers. Optional `mod` → returns `(base ** exp) % mod`. |
| `round(number, ndigits=None)` | `number`: value to round. `ndigits`: int — decimals to keep (negative rounds left of the point). Omitted → nearest int. |
| `min(iterable, *, key=None, default=...)` or `min(a, b, *args)` | An iterable **or** 2+ positional args. `key`: function to rank by. `default`: returned if iterable is empty. |
| `max(...)` | Same shape as `min()`, returns the largest. |
| `sum(iterable, /, start=0)` | `iterable`: numbers to add. `start`: value to begin from (default 0). |

```python
# abs() — absolute value
print(abs(-5), abs(3.2), abs(-7.0))

# divmod() — (quotient, remainder)
print(divmod(17, 5), divmod(9, 3), divmod(20, 6))

# pow() — x**y  (3rd arg = modulus)
print(pow(2, 3), pow(5, 0), pow(2, 10, 1000))

# round() — round to n decimals (negative n rounds left of the point)
print(round(3.14159, 2), round(2.5), round(1234.567, -2))

# min() / max()
print(min(4, 2, 8), max([4, 2, 8]), min("apple", "banana"))

# sum() — optional start value
print(sum([1, 2, 3]), sum([1, 2, 3], 100), sum(range(5)))
```

## 2. Type Conversion / Constructors

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `int(x=0, base=10)` | `x`: a number or string. `base`: only with a string (e.g. `int("1010", 2)`). |
| `float(x=0.0)` | `x`: a number or a numeric string like `"3.14"`, `"1e3"`. |
| `complex(real=0, imag=0)` | Two numbers, **or** a single string like `"4+5j"` (no spaces). |
| `bool(x=False)` | Any object → `True`/`False` by truthiness (0, empty, `None` → `False`). |
| `str(object='', encoding, errors)` | Any object → its readable string form. |
| `bytes(source, encoding, errors)` | A string + encoding, an int (zero-filled length), or an iterable of ints 0–255. |
| `bytearray(...)` | Same args as `bytes()` but **mutable**. |
| `list(iterable)` / `tuple(iterable)` / `set(iterable)` / `frozenset(iterable)` | One optional iterable to build from. |
| `dict(**kwargs)` / `dict(mapping)` / `dict(iterable_of_pairs)` | Keyword args, another mapping, or `(key, value)` pairs. |

```python
# int() — to integer (optional base)
print(int("42"), int(3.9), int("1010", 2))

# float()
print(float("3.14"), float(7), float("1e3"))

# complex()
print(complex(2, 3), complex("4+5j"), complex(6))

# bool() — truthiness
print(bool(0), bool(""), bool([1, 2]))

# str()
print(str(42), str(3.14), str([1, 2]))

# bytes() / bytearray()
print(bytes("hi", "utf-8"), bytes([65, 66, 67]), bytearray("ok", "utf-8"))

# collection constructors
print(list("abc"), tuple([1, 2]), set([1, 1, 2]), dict(a=1, b=2), frozenset([1, 2]))
```

## 3. Base & Character Conversion

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `bin(x)` | `x`: an integer. Returns a `'0b...'` string. |
| `oct(x)` | `x`: an integer. Returns a `'0o...'` string. |
| `hex(x)` | `x`: an integer. Returns a `'0x...'` string. |
| `chr(i)` | `i`: an int code point (0–1,114,111). Returns the matching character. |
| `ord(c)` | `c`: a **single** character string. Returns its int code point. |
| `ascii(object)` | Any object → `repr()` with non-ASCII chars escaped (`\xNN`, `\uNNNN`). |

```python
# bin() / oct() / hex() — integer to base string
print(bin(5), oct(8), hex(255))

# chr() — code point -> character
print(chr(65), chr(97), chr(9731))   # A, a, ☃

# ord() — character -> code point
print(ord("A"), ord("a"), ord("☃"))

# ascii() — printable repr, escaping non-ASCII
print(ascii("café"), ascii("naïve"), ascii("hi"))
```

## 4. Iterables & Sequences

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `len(s)` | `s`: a sequence or collection (string, list, dict, set, ...). Returns item count. |
| `range(stop)` or `range(start, stop, step=1)` | Integers only. `stop` is exclusive; `step` can be negative. |
| `enumerate(iterable, start=0)` | An iterable + optional starting index. Yields `(index, item)`. |
| `zip(*iterables, strict=False)` | Two or more iterables. Stops at the shortest (unless `strict=True`). |
| `sorted(iterable, *, key=None, reverse=False)` | `key`: function to sort by. `reverse`: descending if `True`. Returns a new list. |
| `reversed(seq)` | `seq`: a sequence with a known length. Returns a reverse iterator. |
| `iter(obj)` | An iterable (or `iter(callable, sentinel)`). Returns an iterator. |
| `next(iterator, default=...)` | An iterator. Optional `default` returned instead of raising `StopIteration`. |
| `slice(stop)` or `slice(start, stop, step=None)` | Integers/`None`. Builds a reusable slice object. |
| `filter(function, iterable)` | `function` returning truthy/falsy (or `None` to keep truthy items). |
| `map(function, *iterables)` | A function + one or more iterables fed in parallel as its arguments. |
| `all(iterable)` / `any(iterable)` | One iterable. `all` → True if every item is truthy; `any` → True if at least one is. |

```python
nums = [10, 20, 30]

# len()
print(len("hello"), len(nums), len({"a": 1, "b": 2}))

# range()
print(list(range(5)), list(range(2, 10, 2)), list(range(10, 0, -2)))

# enumerate() — (index, item)
print(list(enumerate(["a", "b"])), list(enumerate(["x", "y"], start=1)))

# zip() — pair up iterables
print(list(zip([1, 2], ["a", "b"])), dict(zip(["x", "y"], [10, 20])))

# sorted() — returns a new sorted list
print(sorted([3, 1, 2]), sorted("dbca"), sorted([3, 1, 2], reverse=True))

# reversed()
print(list(reversed([1, 2, 3])), list(reversed("abc")))

# iter() / next()
it = iter([100, 200])
print(next(it), next(it))

# slice() — reusable slice object
s = slice(1, 4)
print("abcdef"[s], [0, 1, 2, 3, 4, 5][slice(0, 6, 2)])

# filter() — keep items where the test is True
print(list(filter(lambda x: x > 15, nums)), list(filter(None, [0, 1, 2, 0])))

# map() — apply a function to each item
print(list(map(str.upper, ["a", "b"])), list(map(lambda x: x * 2, nums)))

# all() / any()
print(all([True, 1, "x"]), all([1, 0]), any([0, "", 5]))
```

## 5. Input / Output & Formatting

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)` | Any number of objects. `sep`: between them. `end`: after the last. `file`: where to write. |
| `input(prompt='')` | Optional prompt string. Returns the typed line as a **string** (always). |
| `open(file, mode='r', encoding=None, ...)` | `file`: path. `mode`: `'r'`/`'w'`/`'a'`/`'b'` etc. Returns a file object. |
| `format(value, format_spec='')` | `value` + a format-spec string (`'.2f'`, `'x'`, `','`, `'>10'`, ...). |
| `repr(object)` | Any object. Returns an unambiguous string (ideally valid Python). |

```python
# print() — sep and end control the output
print("a", "b", "c", sep="-")
print("no newline", end=" >>> ")
print("then this")

# format() — apply a format spec
print(format(3.14159, ".2f"), format(255, "x"), format(1000000, ","))

# repr() — unambiguous representation (note the quotes/escapes)
print(repr("hi\n"), repr([1, 2]), repr(3.14))

# input() and open() are interactive / file-based — shown but not run:
#   name = input("Your name: ")
#   with open("file.txt", "w") as f:
#       f.write("data")
```

## 6. Objects, Classes & Attributes

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `type(object)` | Any object → its class. (3-arg form `type(name, bases, dict)` builds a new class.) |
| `isinstance(obj, classinfo)` | `obj` + a class **or tuple of classes**. Returns `True`/`False`. |
| `issubclass(cls, classinfo)` | A **class** + a class or tuple of classes. |
| `getattr(obj, name, default=...)` | `name`: attribute name as a string. Optional `default` if missing. |
| `setattr(obj, name, value)` | Sets `obj.name = value` (name as a string). |
| `hasattr(obj, name)` | `name` as a string → `True`/`False`. |
| `delattr(obj, name)` | Deletes the named attribute. |
| `vars(object=...)` | An object → its `__dict__`. No arg → like `locals()`. |
| `dir(object=...)` | An object → list of valid names. No arg → names in current scope. |
| `id(object)` | Any object → unique integer identity for its lifetime. |
| `hash(object)` | A **hashable** object (immutable) → its int hash. |
| `callable(object)` | Any object → `True` if it can be called like `obj()`. |

```python
class Dog:
    species = "canine"
    def __init__(self, name):
        self.name = name

d = Dog("Rex")

# type()
print(type(42), type("hi"), type(d))

# isinstance() / issubclass()
print(isinstance(d, Dog), isinstance(42, (int, float)), issubclass(bool, int))

# getattr / setattr / hasattr / delattr
print(getattr(d, "name"), hasattr(d, "name"))
setattr(d, "age", 3)
print(d.age)
delattr(d, "age")
print(hasattr(d, "age"))

# vars() — the object's __dict__ ;  dir() — available names
print(vars(d))
print("species" in dir(d))

# id() / hash() / callable()
print(type(id(d)), hash("abc") == hash("abc"), callable(d), callable(len))
```

## 7. Decorators / Descriptors (used inside classes)

**Syntax — what each expects:** these are normally applied with the `@` decorator syntax above a method.

| Signature | Arguments |
|---|---|
| `property(fget=None, fset=None, fdel=None, doc=None)` | Getter/setter/deleter functions. As `@property` it wraps a getter; add `@name.setter` for writes. |
| `staticmethod(function)` | Wraps a method that takes **no** `self`/`cls`. Used as `@staticmethod`. |
| `classmethod(function)` | Wraps a method whose first arg is the class `cls`. Used as `@classmethod`. |

```python
class Circle:
    def __init__(self, r):
        self._r = r

    @property                 # managed attribute (accessed like data, no ())
    def area(self):
        return 3.14159 * self._r ** 2

    @staticmethod             # no self / cls — just lives in the namespace
    def info():
        return "A round shape"

    @classmethod              # receives the class as cls
    def unit(cls):
        return cls(1)

c = Circle(2)
print(round(c.area, 2))       # property -> used like an attribute
print(Circle.info())          # staticmethod
print(Circle.unit().area)     # classmethod builds an instance
```

## 8. Scope & Dynamic Execution

**Syntax — what each expects:**

| Signature | Arguments |
|---|---|
| `globals()` | No args. Returns the module's global namespace as a dict. |
| `locals()` | No args. Returns the current local namespace as a dict. |
| `eval(expression, globals=None, locals=None)` | A string/code object that is a **single expression**. Optional namespace dicts. Returns its value. |
| `exec(object, globals=None, locals=None)` | A string/code object of **statements**. Returns `None`. |
| `compile(source, filename, mode, ...)` | `source`: code string. `filename`: label for errors. `mode`: `'eval'`, `'exec'`, or `'single'`. |

⚠️ `eval`/`exec`/`compile` run arbitrary code — never feed them untrusted input.

```python
x = 10

# globals() / locals()
def demo():
    y = 5
    return ("y" in locals(), "x" in globals())
print(demo())

# eval() — evaluates ONE expression and returns its value
print(eval("2 + 3 * 4"), eval("x + 5"))

# exec() — runs statements (returns None)
exec("z = 99")
print(z)

# compile() — source -> code object, then run with eval/exec
code = compile("7 ** 2", "<string>", "eval")
print(eval(code))

# ⚠️ Never run eval()/exec() on untrusted input — it executes arbitrary code.
```
